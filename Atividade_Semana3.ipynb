Importando Bibliotecas

Importando o Dataset e organizando os dados utilizados

```python
# Dataset importado do link: https://www.kaggle.com/datasets/jealousleopard/goodreadsbooks
```

```python
df = pd.read_csv('archive/books.csv', engine= 'python', sep=',', quotechar='"', error_bad_lines=False)
df.head(10)
```

```python
# Padronizando o nome da autora
df["authors"]= df["authors"].str.replace("J.K. Rowling/Mary GrandPré", "J.K. Rowling", case = False)
df
```

```python
# Verificando o tamanho do dataset
df.shape
```

```python
# Definindo o index do Dataset
df.index = df['bookID']
df
```

```python
# Identificando as colunas e o tipo de cada coluna trabalhada
df.info()
```

```python
# Verificando a existência de dados faltantes
df.isna().sum()
```

```python
# Entendendo como cada linha do dataset se comporta
df.loc[df['average_rating'].idxmax()]
```

```python
# Verificando os itens únicos que aparecem no dataset
df.nunique()
```

A verificação de itens únicos se dá necessária, visto que, vários autores podem aparecer na lista com livros em outras linguagens, sendo considerado um novo item de avaliação segundo o Goodreads.

```python
# Identificando os autores com mais livros no Dataset
autores_com_mais_livros = df.groupby('authors')['title'].count().reset_index().sort_values('title', ascending=False).head(10).set_index('authors')
autores_com_mais_livros 
```

```python
# Verificando o número de títulos por editora
editoras_com_mais_livros = df.groupby('publisher')['title'].count().reset_index().sort_values('title', ascending=False).head(10).set_index('publisher')
editoras_com_mais_livros 
```

```python
livros_com_mais_reviews = df.sort_values('text_reviews_count', ascending = False).head(10).set_index('title')
livros_com_mais_reviews 
```

Análise do Dataset

```python
# Dropando colunas que não serão necessárias na análise
df = df.drop(labels=['isbn','isbn13'], axis=1)
df
```

```python
# Contabilizando o número de votos para a definição da nota
quantidade_review = df.groupby('average_rating').agg({'ratings_count': np.sum}).reset_index().tail()
quantidade_review
```

```python
# Identificando as 15 notas médias que mais foram atribuídas aos livros
medias = df.filter(items=['average_rating', 'ratings_count']).groupby('average_rating').count().sort_values('ratings_count', ascending=False).head(15)
medias
```

Podemos identificar, que a nota média com maior número de incidência no dataset é a nota 4.00

```python
# Definindo variavéis para livros, notas e editoras
livros = df['title'].value_counts()[:20]
notas = df.average_rating[:20]
editoras = df['publisher'].value_counts()[:20]
```

```python
# Livros que aparecem com maior frequência no Dataset
plt.figure(figsize=(20,15))
sns.barplot(x = livros, y = livros.index, palette='deep')
plt.title("Títulos Mais Recorrentes")
plt.xlabel("Número de Ocorrências")
plt.ylabel("Títulos")
plt.show()
```

```python
# Editoras que aparecem com maior frequência no Dataset
plt.figure(figsize=(20,15))
sns.barplot(x = editoras, y = editoras.index, palette='icefire_r')
plt.title("Editoras Mais Recorrentes")
plt.xlabel("Número de Ocorrências")
plt.ylabel("Editora")
plt.show()
```

Provavelmente os livros que possuem uma ocorrência alta, foram livros lançados em várias línguas diferentes, assim virando um novo item para o Dataset.

```python
# Número de títulos por linguagem
plt.figure(figsize=(22,10))
ax = df.groupby('language_code')['title'].count().plot.bar()
plt.title('Língua dos itens')
plt.xlabel("Código da linguagem")
plt.xticks(fontsize = 15)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x()-0.3, p.get_height()+100))
```

Fato curioso desse gráfico, é a separação da língua inglesa em 4 categorias diferentes.

```python
# Verificando os autores que mais aparecem no Dataset, e quantos livros cada autor possui no mesmo.
plt.figure(figsize=(15,10))
ax = sns.barplot(autores_com_mais_livros['title'], autores_com_mais_livros.index, palette='icefire_r')
ax.set_title("Os 10 autores com mais livros")
ax.set_xlabel("Número total de livros")
ax.set_ylabel("Autores")
for i in ax.patches:
    ax.text(i.get_width()+.3, i.get_y()+0.5, str(round(i.get_width())), fontsize = 10)
```

```python
# Verificando a ocorrência das notas
plt.figure(figsize=(15,10))
plt.title('Ocorrência das Notas', fontsize= 20)
sns.histplot(df['average_rating'])
```

Podemos verificar a alta incidência de notas entre 3.5 a 4.5

```python
# Verificando o número de reviews por livro
plt.figure(figsize=(15,10))
ax = sns.barplot(livros_com_mais_reviews['text_reviews_count'], livros_com_mais_reviews.index, palette='icefire_r')
ax.set_xlabel("Número de reviews")
ax.set_ylabel("Título")
for i in ax.patches:
    ax.text(i.get_width()+2, i.get_y()+0.5,str(round(i.get_width())), fontsize=10,color='black')
plt.show()
```

```python
# Autores com notas acima de 4.3 por número de livros
autores_com_maiores_notas = df[df['average_rating']>=4.3]
autores_com_maiores_notas = autores_com_maiores_notas.groupby('authors')['title'].count().reset_index().sort_values('title', ascending = False).head(10).set_index('authors')
autores_com_maiores_notas
```

```python
# Gráfico com os autores com notas acima de 4.3 de média
plt.figure(figsize=(15,10))
ax = sns.barplot(autores_com_maiores_notas['title'], autores_com_maiores_notas.index, palette='icefire_r')
ax.set_xlabel("Número de Livros")
ax.set_ylabel("Autores")
for i in ax.patches:
    ax.text(i.get_width()+.3, i.get_y()+0.5, str(round(i.get_width())), fontsize = 10, color = 'k')
```
